```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1 layer model

# use this model to test the hypothesis that global warming is fully explainable by an increase in solar forcing
sigma = 5.67e-8
def pred_temp(albedo, solar_irradiance = None):
    return ( (solar_irradiance * (1 - albedo)) / (4 * sigma)) ** (1/4)
```

```python
year = np.array([1900, 1950, 2000])
s0 = np.array([1365, 1366.5, 1368])
t_anom = np.array([-.4, 0, .4])
predicted_temps = np.array([])
for s in s0: 
    temp_predicted = pred_temp(0.33, s)
    predicted_temps = np.append(predicted_temps, temp_predicted)

# baseline of temperature
temp_baseline = predicted_temps[1]
t_anom_predicted = predicted_temps - temp_baseline
plt.plot(year, t_anom_predicted, label = 'predicted temperature change')
plt.plot(year, t_anom, label = 'observed temperature change ')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.legend()
```

```python
temp_change_predicted = t_anom_predicted[-1] - t_anom_predicted[0]
temp_change_observed = t_anom[-1] - t_anom[0]
print(f'predicted temp change: {temp_change_predicted}')
print(f'observed temp change: {temp_change_observed}')
```
